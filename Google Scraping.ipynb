```python
# Dependencies
import requests
import json

# Google developer API key
from config import gkey
```

```python
target_type = "convenience_store" #check out target types here https://developers.google.com/places/web-service/supported_types#table1
chicago_coords = "41.881832, -87.623177"
radius = 2000

# rewrite params dict

params = {
    "location": chicago_coords,
    "types": target_type,
    "radius": radius,
    "key": gkey
}

# Build URL using the Google Maps API
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# Run request
response = requests.get(base_url, params)

# print the response URL, avoid doing for public GitHub repos in order to avoid exposing key
# print(response.url)

chicago_gasstattions = response.json()

# Print the JSON (pretty printed)
print(json.dumps(chicago_gasstattions, indent=4, sort_keys=True))
```
